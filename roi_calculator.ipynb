```python
class ROI:
    
    def __init__(self, name):
        self.name = name
        
    def get_rental_income(self):
        global income
        income = int(input("\nWhat is the monthly rental income for your property?\n"))
        
    def get_expenses(self):
        expenses = []
        global total_expenses
        taxes = int(input("\nWhat is the monthly tax fee for your property?\n"))
        expenses.append(taxes)
        insurance = int(input("\nWhat is the monthly insurance fee for your property?\n"))
        expenses.append(insurance)
        utilities = input("\nDo you pay the utitlity bills for your property?\n")
        while utilities not in {'yes', 'no'}:
             utilities = input("\nInvalid Answer.\nDo you pay the utitlity bills for your property?\n")
        if utilities == 'yes':
            electric = int(input("\nWhat is the monthly electricity fee?"))
            water = int(input("What is the monthly water fee?"))
            sewer = int(input("What is the monthly sewer fee?"))
            gas = int(input("What is the monthly gas fee?"))
            utility_fees = electric + water + sewer + gas
            expenses.append(utility_fees)
            print(f"\nYour Total UTILITY Expenses: ${utility_fees:.2f}\n")
        hoa = int(input("\nWhat is the monthly HOA fee for your property?\n"))
        expenses.append(hoa)
        property_management = int(input("\nWhat is the property management fee for your property?\n"))
        expenses.append(property_management)
        other_fees = input("\nDo you have any other monthly expenses. If so, enter yes, if no, enter no to continue.\n")
        while other_fees not in {'yes', 'no'}:
            other_fees = input("\nInvalid Answer. Do you have any other monthly expenses. If so, enter yes, if no, enter no to continue.\n")
        if other_fees == 'yes':
            extra_fees = int(input("\nPlease calculate and enter your other fees:\n"))
            expenses.append(extra_fees)
            total_expenses = int(sum(expenses))
            print(f"\nYour Total EXPENSES: ${total_expenses:.2f}\n")
        if other_fees == 'no':
            total_expenses = int(sum(expenses))
            print(f"\nYour Total EXPENSES: ${total_expenses:.2f}\n")
        
    def get_cash_flow(self):
        global cash_flow
        cash_flow = income - total_expenses
        print(f"\nYour Total CASH FLOW: ${cash_flow:.2f}\n")
        
    def get_total_investment(self):
        global total_investment
        total_investment = []
        down_payment = int(input("\nWhat is the down payment for your property?"))
        total_investment.append(down_payment)
        closing_costs = int(input("What is the closing costs for your property?"))
        total_investment.append(closing_costs)
        repair_money = int(input("Did you have to make repairs to your property? If so calculate and enter the amount."))
        total_investment.append(repair_money)
        misc_fees = int(input("Please calculate any other fees invested into your property. If you have no other fees please enter 0."))
        total_investment.append(misc_fees)
        total_investment = sum(total_investment)
        print(f"\nYour Total INVESTMENT for your property: ${total_investment:.2f}\n")
        
    def get_ROI(self):
        annual_cash_flow = cash_flow * 12
        decimal = annual_cash_flow / total_investment
        new_decimal = decimal * 100
        print(f"ROI = {new_decimal}%")
    
def run_ROI():
    name = input("\nWhat is your name?\n")
    print(f"\nHi {name.title()}!\n\nWe look forward to helping you calculate the ROI for your property!\n")
    print("\nFirst Step!")
    my_ROI = ROI(name)
    my_ROI.get_rental_income()
    print("\nSecond Step!")
    my_ROI.get_expenses()
    my_ROI.get_cash_flow()
    print("\nThird Step!")
    my_ROI.get_total_investment()
    my_ROI.get_ROI()
    print("\nAll Done!")
    
    
        
run_ROI()
      
```
